```python
def get_bounding_box(image_w,image_h,x_center, y_center, w, h):
        #coordinates of the bounding box
        x_center_pixel= x_center * image_w
        y_center_pixel= y_center * image_h
        half_w=(w*image_w)/2
        half_h=(h*image_h)/2


        xmin=int(x_center_pixel-half_w)
        xmax=int(x_center_pixel+half_w)
        ymin=int(y_center_pixel-half_h)
        ymax=int(y_center_pixel+half_h)

        return(xmin,xmax,ymin,ymax)
    
def crop(imagen,label):
    import cv2
    image = cv2.imread(imagen)
    image_h,image_w,_=image.shape
    v=get_bounding_box(image_w,image_h,float(label[0]), float(label[1]), float(label[2]), float(label[3]))
    y,h,x,w=v
    crop_image = image[x:w, y:h]
    cv2.imshow('Cropped', crop_image)
    cv2.waitKey(0)
    return(crop_image)

def guardar_imagen(imagen, directorio_destino, nombre_archivo):

    # Guardar la imagen en el directorio especificado
    if not os.path.exists(directorio_destino):
        os.makedirs(directorio_destino)
    
    ruta_destino = os.path.join(directorio_destino, nombre_archivo)
    plt.imsave(ruta_destino, imagen)  # Guardar la imagen usando plt.imsave()
```

```python
#M√âTRICAS
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Ejemplo de etiquetas reales y predicciones
y_true = [1, 0, 1, 1, 0, 1, 0, 0, 1, 0]
y_pred = [1, 0, 1, 1, 0, 0, 1, 0, 1, 0]

# Exactitud (Accuracy)
accuracy = accuracy_score(y_true, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Precisi√≥n (Precision)
precision = precision_score(y_true, y_pred)
print(f'Precision: {precision:.2f}')

# Recall (Sensibilidad)
recall = recall_score(y_true, y_pred)
print(f'Recall: {recall:.2f}')

# F1-Score
f1 = f1_score(y_true, y_pred)
print(f'F1-Score: {f1:.2f}')

# Matriz de confusi√≥n
conf_matrix = confusion_matrix(y_true, y_pred)
print('Matriz de Confusi√≥n:')
print(conf_matrix)

# Curva ROC y √Årea bajo la curva (AUC-ROC)
fpr, tpr, thresholds = roc_curve(y_true, y_pred)
auc_roc = roc_auc_score(y_true, y_pred)

# Visualizaci√≥n de la Curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC = {auc_roc:.2f}')
plt.plot([0, 1], [0, 1], 'k--', label='Random')
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curva ROC')
plt.legend()
plt.show()
```
